```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
import astropy.constants as const

# retina
%config InlineBackend.figure_format='retina'
```

```python
import seaborn

seaborn.set_theme(context="notebook", style="ticks")
```

```python
from gcfit.util.units import angular_width

# enable pc to arcsec equivalencies
u.set_enabled_equivalencies(angular_width(D=5.43 * u.kpc))
```

```python
def vesc(r, cluster=80 * u.km / u.s, MBH=8000 * u.M_sun):
    """
    Calculate the escape velocity profile of a BH given a local cluster escape velocity.

    Parameters
    ----------
    r : Quantity
        Distance from the BH
    cluster : Quantity
        Local cluster escape velocity
    MBH : Quantity
        Mass of the BH

    Returns
    -------
    Quantity
        Escape velocity at distance r
    """

    r = r.to(u.pc)

    vesc_BH = np.sqrt((2 * const.G * MBH) / r)

    # print(f"vesc_BH = {vesc_BH.to(u.km/u.s)}")
    return np.sqrt(cluster**2 + vesc_BH**2)
```

```python
rs = np.linspace(1e-1, 1e2, 10000) * u.arcsec
rs
```

```python
# sanity check
vescs = vesc(rs, cluster=80 * u.km / u.s, MBH=8000 * u.M_sun)


plt.figure()

plt.plot(rs, vescs, label="vesc")

plt.xlabel("r ['']")
plt.ylabel("v [km/s]")
plt.xscale("log")

plt.axhline(80, color="k", linestyle="--", label="cluster vesc")

plt.legend()
```

Now with fast stars

```python
# 2d velocities:

# Measured v2D A: (113.01.1) km s1 B: (66.64.1) km s1 C: (94.91.7) km s1 D: (77.92.0) km s1 E: (69.60.8) km s1 F: (67.40.4) km s1 G: (66.21.9) km s1

v2D = np.array([113.0, 66.6, 94.9, 77.9, 69.6, 67.4, 66.2]) * u.km / u.s
v2D_err = np.array([1.1, 4.1, 1.7, 2.0, 0.8, 0.4, 1.9]) * u.km / u.s


# Acceleration R.A. [mas yr-2] -0.00690.0083 (0.8) 0.07020.0239 (2.9) 0.00280.0333 (0.1) 0.03570.0177 (2.0) 0.00720.0042 (1.7) 0.00520.0038 (1.4)  -0.01970.0267 (0.7)

# Acceleration Dec. [mas yr-2] 0.00850.0098 (0.9) 0.02280.0157 (1.5) -0.00600.0123 (0.5) -0.01940.0162 (1.2) -0.00090.0075 (0.1) -0.00150.0038 (0.4) 0.01730.0170 (1.0)

a_ra = (
    np.array([-0.0069, 0.0702, 0.0028, 0.0357, 0.0072, 0.0052, -0.0197])
    * u.mas
    / u.yr**2
)
a_ra_err = (
    np.array([0.0083, 0.0239, 0.0333, 0.0177, 0.0042, 0.0038, 0.0267]) * u.mas / u.yr**2
)

a_dec = (
    np.array([0.0085, 0.0228, -0.0060, -0.0194, -0.0009, -0.0015, 0.0173])
    * u.mas
    / u.yr**2
)
a_dec_err = (
    np.array([0.0098, 0.0157, 0.0123, 0.0162, 0.0075, 0.0038, 0.0170]) * u.mas / u.yr**2
)


r2d = np.array([0.265, 0.537, 0.870, 0.886, 1.333, 2.408, 2.506]) * u.arcsec


r2d_pc = r2d.to(u.pc)


labels = ["A", "B", "C", "D", "E", "F", "G"]
```

```python
from astropy.table import Table
```

```python
stars = Table(
    {
        "label": labels,
        "r2d_pc": r2d_pc,
        "r2d_as": r2d,
        "v2D": v2D,
        "v2D_err": v2D_err,
        "a_ra": a_ra,
        "a_ra_err": a_ra_err,
        "a_dec": a_dec,
        "a_dec_err": a_dec_err,
    }
)
```

```python
stars
```

```python
# add above to plot


vescs = vesc(rs, cluster=62 * u.km / u.s, MBH=8200 * u.M_sun)

plt.figure(figsize=(10, 5))

plt.xlabel("r [arcsec] (AvdM10)")
plt.ylabel("v [km/s]")

plt.xscale("log")

plt.ylim(58, 145)

plt.axhline(62, color="k", linestyle="--")

plt.plot(rs, vescs, label="vesc")

colors = [f"C{i+1}" for i in range(len(stars))]
for i in range(len(stars)):
    plt.errorbar(
        x=stars["r2d_as"][i],
        y=stars["v2D"][i],
        yerr=stars["v2D_err"][i],
        marker="o",
        label=stars["label"][i],
        color=colors[i],
    )

plt.legend()
plt.grid()
```

This looks pretty much like their plot, now find minimum BH mass

```python
from gcfit.util.probabilities import gaussian_likelihood
```

```python
# assume a cluster escape velocity of 80 km/s
def log_like(MBH):

    vescs = vesc(rs, cluster=80 * u.km / u.s, MBH=MBH * u.M_sun)

    vescs_interp = np.interp(x=stars["mmc_rad"], xp=rs, fp=vescs)

    # enforce that the escape velocity is greater than the 2D velocity within errors
    if any(vescs_interp < stars["v2D"] - stars["v2D_err"]):
        return -np.inf

    ll = gaussian_likelihood(
        X_data=stars["v2D"].data, X_model=vescs_interp, err=stars["v2D_err"].data
    )
    if np.isfinite(ll):
        return ll
    return -np.inf
```

```python
log_like(10000)
```

```python
import emcee

pos0 = [10000]

pos = pos0 + 1e-4 * np.random.randn(16, 1)
nwalkers, ndim = pos.shape

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_like)
sampler.run_mcmc(pos, 2000, progress=True)
```

```python
flat_samples = sampler.get_chain(discard=500, thin=5, flat=True)
fig, axes = plt.subplots(1, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
labels = [
    "MBH",
]
for i in range(ndim):
    ax = axes
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes.set_xlabel("step number")
```

```python
from corner import corner

flat_samples = sampler.get_chain(discard=500, thin=5, flat=True)
fig = plt.figure(figsize=(8, 6))
fig = corner(
    flat_samples,
    labels=["MBH"],
    quantiles=[0.16, 0.5, 0.84],
    show_titles=True,
    title_kwargs={"fontsize": 18},
    fig=fig,
)
fig.tight_layout()
```

For a lower limit we only actually care what the minimum allowed mass is.

```python
MBH_min = np.min(flat_samples.flatten())
MBH_min
```

```python
# add above to plot


vescs = vesc(rs, cluster=80 * u.km / u.s, MBH=MBH_min * u.M_sun)

plt.figure(figsize=(10, 5))

plt.xlabel("r 2D [arcsec] (Minimum mass center)")
plt.ylabel("v 2D [km/s]")

plt.xscale("log")

plt.ylim(50, 150)

plt.axhline(80, color="k", linestyle="--", label="cluster vesc")

plt.plot(rs, vescs, label=f"MBH={MBH_min:.0f}")

colors = [f"C{i+1}" for i in range(len(stars))]
for i in range(len(stars)):
    plt.errorbar(
        x=stars["mmc_rad"][i],
        y=stars["v2D"][i],
        yerr=stars["v2D_err"][i],
        marker="o",
        label=stars["label"][i],
        color=colors[i],
    )
plt.xlim(1e-1, 1e2)
plt.legend()
plt.grid()
```

So using their center, we can recover the 8200 solar mass lower limit when we use 62 km/s as the escape velocity.

Raising the escape velocity to 80 km/s, we get a lower limit of 5200 solar masses.

Quick look at accelerations

```python
# look at acceleration now

# total acceleration
stars["a_tot"] = np.sqrt(stars["a_ra"] ** 2 + stars["a_dec"] ** 2)
stars["a_tot_err"] = np.sqrt(stars["a_ra_err"] ** 2 + stars["a_dec_err"] ** 2)
```

```python
# plots acceleration vs distance from minimum mass center

plt.figure(figsize=(10, 5))

plt.xlabel("r 2D [arcsec] (Minimum mass center)")
plt.ylabel("a [mas/yr^2]")
plt.xscale("log")

colors = [f"C{i+1}" for i in range(len(stars))]

for i in range(len(stars)):
    plt.errorbar(
        x=stars["mmc_rad"][i],
        y=stars["a_tot"][i],
        yerr=stars["a_tot_err"][i],
        marker="o",
        label=stars["label"][i],
        color=colors[i],
    )

plt.legend()
plt.grid()
```

```python
# RA dec plot with acceleration vectors

plt.figure(figsize=(10, 10))

plt.xlabel("RA [deg]")
plt.ylabel("Dec [deg]")
plt.gca().invert_xaxis()

colors = [f"C{i+1}" for i in range(len(stars))]

# draw acceleration vectors
for i in range(len(stars)):
    plt.arrow(
        x=stars["RA"][i],
        y=stars["Dec"][i],
        dx=stars["a_ra"][i] * 1e-3,
        dy=stars["a_dec"][i] * 1e-3,
        color=colors[i],
        width=0.000005,
        label=stars["label"][i],
    )


# draw the minimum mass center
plt.scatter(mmc.ra, mmc.dec, color="k", label="Minimum mass center")

plt.legend()
```
